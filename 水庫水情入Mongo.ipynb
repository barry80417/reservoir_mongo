### 讀取水利署水庫水情資料

```python
import requests
url ='https://fhyv.wra.gov.tw/FhyWeb/v1/Api/Reservoir/Visual?$format=JSON'
r = requests.get(url)
response = r.text
```

### 將json讀成data再做成dataframe

```python
import json
import pandas as pd
data = json.loads(response)
df = pd.json_normalize(data,record_path = ['ReservoirRealTimeInfos'])
df
```

### 讀取Mongo伺服器

```python
import pymongo
client = pymongo.MongoClient('localhost', 27017)
```

### 選取資料庫

```python
database = client["mongo-sample"]
collection = database["reservoir"]

records = df.to_dict('records') 
collection.insert_many(records)
```
